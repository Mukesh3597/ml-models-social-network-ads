```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r"C:\Users\mukes\Downloads\Social_Network_Ads.csv")
```

```python
df.shape
```

```python
x = df.iloc[:,:-1]
y =df["Purchased"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x)
x1=sc.transform(x)
```

```python
x = pd.DataFrame(x1,columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
dt =KNeighborsClassifier(n_neighbors=5)
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100
```

```python
(82.1875, 90.0)
```

```python
dt.predict([[19,19000]])
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel="rbf")
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100
```

```python
sv.predict([[46,41000]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

```python
sns.scatterplot(x="Age", y="Purchased", data=df)
plt.plot(df["Age"],sv.predict(x),color="red")
plt.show()
```
